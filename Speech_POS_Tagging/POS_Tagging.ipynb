```python
nlp = spacy.load("en_core_web_sm")
```

```python
nlp.pipe_names
```

```python
doc = nlp("Elon flew mars yesterday. He cared biryani masala with him")

for token in doc:
    print(token, " | ", token.pos_, " | ", spacy.explain(token.pos_), " | ", token.tag_, " | ", spacy.explain(token.tag_))
```
